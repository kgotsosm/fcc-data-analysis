# Sea Level Predictor

As part of the Data Analysis course on Free Code Camp, I built a sea level predictor that analysed the dataset called: "Global Average Absolute Sea Level Change, 1880-2014 from the US Environmental Protection Agency using data from CSIRO, 2015; NOAA, 2015". The purpose of the exercise was to use Python libraries to predict the sea level up to the year 2050 based on the initial data given. 

### Import Libraries

The libraries used here will be Pandas, Numpy, Matplotlib and Scipy.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

## Function To Create Visualisation

The function below will read the epa-sea-level dataset which is in csv format then prepare it for visualisation. Using the libraries imported, we can then create a scatter plot and create two lines of best fit to extrapolate the potential sea levels of the future from the data.

```python
def draw_plot():
    # Read data from file
    df = pd.read_csv('epa-sea-level.csv')

    # Create scatter plot
    plt.scatter(x=df['Year'], y=df['CSIRO Adjusted Sea Level'])

    # Create first line of best fit
    lineA = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    xA = np.arange(df['Year'].min(),2050,1)
    yA = xA*lineA.slope + lineA.intercept

    plt.plot(xA,yA)

    # Create second line of best fit
    df_2000 = df[df['Year'] >= 2000]

    lineB = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])
    xB = np.arange(2000,2050,1)
    yB = xB*lineB.slope + lineB.intercept

    plt.plot(xB,yB)

    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
    plt.xticks(np.arange(1880, 2050, step=20))

    
    
    # Save plot and return data for testing (DO NOT MODIFY)

    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
draw_plot()
```
